```python
from src.trainer import run_experiment
import os

HF_TOKEN = os.getenv("HF_TOKEN")

# Base config for DistilBERT + SST2
base_config = {
    "task_name": "sst2",
    "model_name": "distilbert",
    "num_epochs": 3,
    "batch_size": 16,
    "learning_rate": 5e-4,
    "wandb_project": "lora-vs-ia3-distilbert",
}
```

```python
# --- LoRA Sweep ---
for r_val in [8, 16]:
    run_config_lora = base_config.copy()
    run_config_lora.update({
        "run_name": f"distilbert_sst2_lora_r{r_val}",
        "peft_method": "lora",
        "r": r_val,
        "lora_alpha": r_val * 2, # Common practice: alpha = 2*r
    })

    print(f"\n--- STARTING LoRA RUN: r={r_val} ---")
    run_experiment(run_config_lora, HF_TOKEN)
    print(f"--- FINISHED LoRA RUN: r={r_val} ---")
```

```python
# --- IA³ Run ---
run_config_ia3 = base_config.copy()
run_config_ia3.update({
    "run_name": "distilbert_sst2_ia3",
    "peft_method": "ia3",
})

print(f"\n--- STARTING IA³ RUN ---")
run_experiment(run_config_ia3, HF_TOKEN)
print(f"--- FINISHED IA³ RUN ---")
```
